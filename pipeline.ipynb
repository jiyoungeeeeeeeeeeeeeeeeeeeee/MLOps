## pipeline

```python
#       
def find_mixed_type_columns(df):
    mixed_type_columns = []
    for col in df.columns:
        if df[col].dtype == 'object':
            try:
                df[col].astype(float)
            except ValueError:
                mixed_type_columns.append(col)
    return mixed_type_columns


#      
mixed_columns = find_mixed_type_columns(data)

#    dtype  
def convert_dtype(X):
    X = X.copy()
    for col in mixed_columns:
        X[col] = X[col].astype(str)
    return X

numeric_col = list(data.select_dtypes(include=['float64', 'int64']).columns)
categoric_col = list(data.select_dtypes(include=['object']).columns)


#   
dtype_transformer = FunctionTransformer(convert_dtype)

numeric_transformer = Pipeline(steps=[
    ('num_imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('PCA' , PCA(n_components = 2))
])

categoric_transformer = Pipeline(steps=[
    ('cat_imputer', SimpleImputer(strategy='most_frequent')),
    ('onehotencoder', OneHotEncoder(handle_unknown='ignore', drop='first', sparse=True))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('dtype', dtype_transformer, mixed_columns),  # dtype  
        ('num', numeric_transformer, numeric_col),
        ('cat', categoric_transformer, categoric_col)
    ])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])
```

```python
pipeline.fit(X, y)

predictions = pipeline.predict(X)

print(predictions)
```
