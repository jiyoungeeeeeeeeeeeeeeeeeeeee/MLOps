```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline 


pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    LogisticRegression(max_iter=1000, random_state=42)    
)

pipeline.fit(X_train, y_train)
```

```python
# from sklearn.metrics import roc_auc_score


y_valid_preds = pipeline.predict(X_valid)

# 'multi_class' 매개변수를 'ovo' 또는 'ovr'로 설정
#roc_auc = roc_auc_score(y_valid, y_valid_preds, multi_class='ovr')

#print(f"검증 데이터 ROC AUC : {roc_auc :.4f}" )
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_valid , y_valid_preds))

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rfc = RandomForestClassifier(n_estimators= 100, max_depth= 5, random_state = 0)
rfc.fit(X_train , y_train)

predict1 = rfc.predict(X_valid)
print(classification_report(y_valid , predict1))
```

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter


print(f"원본 데이터 클래스 분포 : {Counter(y_train)}")

rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

print(f"랜덤 언더 샘플링 후 클래스 분포 : {Counter(y_resampled)}" )
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rfc = RandomForestClassifier(n_estimators= 100, max_depth= 3 , random_state = 0)
rfc.fit(X_resampled , y_resampled)

predict1 = rfc.predict(X_valid)
print(classification_report(y_valid , predict1))
```

```python
def print_score(y_true, y_pred, train=True):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    if train:
        print("Train Result:")
    else:
        print("Test Result:")
    
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precision: {precision_score(y_true, y_pred, average='weighted'):.4f}")
    print(f"Recall: {recall_score(y_true, y_pred, average='weighted'):.4f}")
    print(f"F1 Score: {f1_score(y_true, y_pred, average='weighted'):.4f}")
```

```python
y_train_pred = rfc.predict(X_train)
y_valid_pred = rfc.predict(X_valid)

print_score(y_train, y_train_pred, train=True)
print_score(y_valid, y_valid_pred, train=False)
```
