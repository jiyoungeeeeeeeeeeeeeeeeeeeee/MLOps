````python
data['earliest_cr_line'] = pd.to_datetime(data['earliest_cr_line'], format = '%b-%Y')
data['issue_d'] = pd.to_datetime(data['issue_d'],format = '%b-%Y')
````

````python
data.info()
````

````python
import pandas as pd
from datetime import datetime

data['cr_line_period'] = (datetime.now() - data['earliest_cr_line']).dt.days
data['issue_d_period'] = (datetime.now() - data['issue_d']).dt.days
````

````python
data.drop('earliest_cr_line' , axis = 1 , inplace= True)
data.drop('issue_d' , axis= 1 , inplace= True)
````

````python
from sklearn.preprocessing import OrdinalEncoder

ord = OrdinalEncoder()

ord_feature = ['cr_line_period' , 'issue_d_period']

data[ord_feature]= ord.fit_transform(data[ord_feature]) 
````

````python
data = data[data['collections_12_mths_ex_med'] < 7]
````

````python
data['collections_12_mths_ex_med'].value_counts()
````

````python
emp_length_1 = {'< 1 year': 1 , '1 year': 2 , '2 years' : 3 , '3 years': 4 , '4 years' : 5 , '5 years' : 6 , '6 years' : 7 , '7 years': 8 , '8 years':9 ,'9 years': 10 , '10+ years':11}

data['emp_length'] = data['emp_length'].map(emp_length_1)
````

````python
data['grade'].value_counts()
````

````python
grade_1 = {'A' : 0 ,'B' : 1, 'C' : 2, 'D' : 3, 'E' : 4, 'F' : 5, 'G' : 6}

data['grade'] = data['grade'].map(grade_1)
````

````python
data['grade'].value_counts()
````

````python
from sklearn.preprocessing import MinMaxScaler

Mm_feature = ['emp_length', 'grade', 'collections_12_mths_ex_med' , 'cr_line_period','issue_d_period']

Mm = MinMaxScaler()

data[Mm_feature] = Mm.fit_transform(data[Mm_feature])
````

````python
from sklearn.preprocessing import OneHotEncoder

ohe_feature = ['home_ownership'  , 'purpose']

ohe = OneHotEncoder()

ohe_num = ohe.fit_transform(data[ohe_feature])
````

````python
data.drop(ohe_feature, axis = 1 , inplace = True)
````

````python
from sklearn.preprocessing import StandardScaler

num_features = ['loan_amnt','int_rate','installment','annual_inc','dti','open_acc','pub_rec','revol_bal','revol_util','total_acc','mort_acc']

std_saller = StandardScaler()

data[num_features] = std_saller.fit_transform(data[num_features])
````

````python
data.head(5)
````

## y값 변환

````python
data['loan_status'].value_counts()
````

````python
target_1 = {'Fully Paid':0 , 'Current':1 , 'Charged Off':2, 'Late (31-120 days)' : 3 , 'In Grace Period': 4, 'Late (16-30 days)':5, 'Default':6 }
data['loan_status']= data['loan_status'].map(target_1)
````

## 모든 데이터 합치기

````python
from scipy import sparse

data_sparse = sparse.hstack([sparse.csr_matrix(data),
                               ohe_num
                               ],
                               
                               format = 'csr')
````

````python
#data.to_csv('./train.csv', index = False)
````

````python
y = data['loan_status']
````

````python
data = data.drop('loan_status' , axis = 1)
````

````python
num_train = len(data)

X_train = data_sparse[:num_train]
X_test = data_sparse[num_train :]
````

````python
from sklearn.model_selection import train_test_split

X_train , X_valid, y_train, y_valid = train_test_split(X_train, y, 
                                                       test_size=0.2, 
                                                       random_state=42 , 
                                                       stratify = y)
````

## 모델 훈련


````python
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline 


pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    LogisticRegression(max_iter=1000, random_state=42)    
)

pipeline.fit(X_train, y_train)
````

````python
# from sklearn.metrics import roc_auc_score


y_valid_preds = pipeline.predict(X_valid)

# 'multi_class' 매개변수를 'ovo' 또는 'ovr'로 설정
#roc_auc = roc_auc_score(y_valid, y_valid_preds, multi_class='ovr')

#print(f"검증 데이터 ROC AUC : {roc_auc :.4f}" )
````

````python
from sklearn.metrics import classification_report

print(classification_report(y_valid , y_valid_preds))

````

````python
y.value_counts(sort= False)
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rfc = RandomForestClassifier(n_estimators= 100, max_depth= 5, random_state = 0)
rfc.fit(X_train , y_train)

predict1 = rfc.predict(X_valid)
print(classification_report(y_valid , predict1))
````

train만 oversampling 실행



````python
y_train.value_counts()
````

````python
from imblearn.under_sampling import OneSidedSelection as OSS 
OSS_X_train , OSS_y_train = OSS(random_state = 42).fit_resample(X_train , y_train)
OSS_y_train.value_counts()
````

````python
import scipy as sp
````

````python
oss_4C_train = sp.hstack((OSS_X_train, OSS_y_train))
````

````python
type(oss_4C_train)
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rfc = RandomForestClassifier(n_estimators= 100, max_depth= 5 , random_state = 0)
rfc.fit(OSS_X_train , OSS_y_train)

predict_oss = rfc.predict(X_valid)
print(classification_report(y_valid , predict_oss))
````
