```python
def summary(df):
    sum = pd.DataFrame(df.dtypes, columns=['dtypes'])
    sum[' '] = df.isna().sum()
    sum[' '] = (df.isna().sum())/len(df)
    sum[' '] = df.nunique().values
    sum[' '] = df.count().values


    return sum

summary(data).style.background_gradient(cmap='Blues')
```
